```python
import tkinter
from tkinter import Entry
from functools import partial

def call_result1(label_result, n1, n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(n1)+int(n2)
    label_result.config(text="Result is %d" % result)
    return
def call_result2(label_result, n1, n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1)-int(num2)
    label_result.config(text="Result is %d" % result)
    return
def call_result2(label_result, n1, n2):
    num1 = (n1.get())
    num2 = (n2.get())
    result = int(num1)*int(num2)
    label_result.config(text="Result is %d" % result)
    return    


root = tkinter.Tk()
root.geometry('400x200+100+200')
root.title('Simple Calculator')
number1 = tkinter.StringVar()
number2 = tkinter.StringVar()
labelTitle = tkinter.Label(root, text="Simple Calculator").grid(row=0, column=2)
labelNum1 = tkinter.Label(root, text="Enter 1st number, A").grid(row=1, column=0)
labelNum2 = tkinter.Label(root, text="Enter 2nd number, B").grid(row=2, column=0)
labelResult = tkinter.Label(root)
labelResult.grid(row=7, column=2)
entryNum1 = tkinter.Entry(root, textvariable=number1).grid(row=1, column=2)
entryNum2 = tkinter.Entry(root, textvariable=number2).grid(row=2, column=2)
call_result1 = partial(call_result1, labelResult, number1, number2)
call_result2 = partial(call_result2, labelResult, number1, number2)
buttonCal = tkinter.Button(root, text="A+B", command=call_result1).grid(row=3, column=0)
buttonCal = tkinter.Button(root, text="A-B", command=call_result2).grid(row=3, column=1)
root.mainloop()
```
